### What is the distribution  of preferred foot?


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
pd.pandas.set_option('display.max_columns',None)
df=pd.read_csv("Footballer.csv")
df.head(5)
```

```python
features = df.columns

[feature for feature in features if df[feature].isnull().sum()>0]
```

```python
$calculating the percentage of null values for the dataset
for feature in features:
  if df[feature].isnull().sum() > 0 :
    print(feature,"\t",round(df[feature].isnull().mean()*100,2))
```

### How can we visualize the age distribution of players?

```python
sns.displot(x=df['Age'])
sns.displot(x=df['Age'],kind='kde')
```

Most of the Players are in between age of 25 - 35

### What is the international reputation of left and right foot?

```python
plt.figure(figsize=(12,8))
sns.countplot(x=df['International Reputation'],hue=df['Preferred Foot'])
pd.crosstab(df['International Reputation'],df['Preferred Foot'])
```

Above visualization shows the Internaltional Reputation in basis of Preferred Foot

### Age distribution in top countries

```python
top_countries=df['Nationality'].value_counts().index.tolist()[0:10]


for i in top_countries:
  Nationality = df[df['Nationality'] == i ]['Nationality']
  Age = df[df['Nationality'] == i]['Age']
  sns.displot(x=Age,hue=Nationality,kind='kde')
```

In top 10 countries, we can see most of the Players are from 25 - 35 years of Age

### What is the distribution of overall score of players from top countries?

```python
for i in top_countries:
  Nationality = df[df['Nationality'] == i ]['Nationality']
  over_all_score = df[df['Nationality'] == i ]['Overall']
  sns.displot(x=over_all_score,hue=Nationality,kind='kde')
```

In top 10 countries, Most of the Players have hit 70 - 75 goals.
